```python
# Part 1:
# Giving a 2D list of "." and "@"
# "@" represents toilet paper and "." represents empty space
# Must find "@" touching less than 4 "@" in all 8 adjacent directions


def solve(file_name):
    total = 0
    grid = []
    with open(file_name, "r") as f:
        grid = [list(line.strip()) for line in f.readlines()]
    # pad the map
    padMap = [
        ["." for _ in range(len(grid[0]) + 2)],
        *[[".", *row, "."] for row in grid],
        ["." for _ in range(len(grid[0]) + 2)],
    ]
    for i in range(1, len(padMap) - 1):
        for j in range(1, len(padMap[0]) - 1):
            # count adjacent "@"
            if padMap[i][j] == "@":
                count = 0
                for di in [-1, 0, 1]:
                    for dj in [-1, 0, 1]:
                        if di == 0 and dj == 0:
                            continue
                        if padMap[i + di][j + dj] == "@":
                            count += 1
                if count < 4:
                    total += 1
    return total


print(f"Part 1 - test: {solve('./data/day4-test.txt')}")
print(f"Part 1 - full data: {solve('./data/day4-data.txt')}")
```

```python
# Part 2:
# Now, we determine the same thing, but include newly available toilet paper
# once rolls are removed from the grid


def solve2(file_name):
    total = 0
    grid = []
    with open(file_name, "r") as f:
        grid = [list(line.strip()) for line in f.readlines()]
    # pad the map
    padMap = [
        ["." for _ in range(len(grid[0]) + 2)],
        *[[".", *row, "."] for row in grid],
        ["." for _ in range(len(grid[0]) + 2)],
    ]
    # Now, we repeat until no more "@" can be removed
    while True:
        current_total = 0
        for i in range(1, len(padMap) - 1):
            for j in range(1, len(padMap[0]) - 1):
                # count adjacent "@"
                if padMap[i][j] == "@":
                    count = 0
                    for di in [-1, 0, 1]:
                        for dj in [-1, 0, 1]:
                            if di == 0 and dj == 0:
                                continue
                            if padMap[i + di][j + dj] == "@":
                                count += 1
                    if count < 4:
                        current_total += 1
                        padMap[i][j] = "x"  # remove the roll
        if current_total == 0:
            break
        total += current_total
    return total


print(f"Part 2 - test: {solve2('./data/day4-test.txt')}")
print(f"Part 2 - full data: {solve2('./data/day4-data.txt')}")
```
