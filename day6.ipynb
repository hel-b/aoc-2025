```python
# Part 1:
# mathematical operations on data columns
# operation indicated in last row
# going to assume only +, -, *, and /
# return sum total of all operations


def solve_part1(file_name):
    data = []
    ops = []
    with open(file_name, "r") as f:
        raw = [line.strip().split() for line in f.read().split("\n")]
        data = [[int(line[i]) for line in raw[:-1]] for i in range(len(raw[0]))]
        ops = raw[-1]
    results = []
    for i, op in enumerate(ops):
        result = data[i][0]
        for val in data[i][1:]:
            if op == "+":
                result += val
            elif op == "-":
                result -= val
            elif op == "*":
                result *= val
            elif op == "/":
                result /= val
        results.append(result)
    return sum(results)


print(f"Part 1 - Test Data: {solve_part1('./data/day6-test.txt')}")
print(f"Part 1 - Real Data: {solve_part1('./data/day6-data.txt')}")
```

```python
# Part 2:
# Now operands are in columns organized right to left
# last row is still operation to perform


def solve_part2(file_name):
    data = []
    ops = []
    with open(file_name, "r") as f:
        raw = f.read().split("\n")
        ops = raw[-1].split()
        raw_vals = ",".join(
            "".join(raw[i][j] for i in range(len(raw[:-1]))) for j in range(len(raw[0]))
        )
        data = [
            [int(val) for val in vals_str.split(",")][::-1]
            for vals_str in raw_vals.split(f",{" " * len(raw[:-1])},")
        ]

    results = []

    for i, op in enumerate(ops):
        result = data[i][0]
        for val in data[i][1:]:
            if op == "+":
                result += val
            elif op == "-":
                result -= val
            elif op == "*":
                result *= val
            elif op == "/":
                result /= val
        results.append(result)
    return sum(results)


print(f"Part 2 - Test Data: {solve_part2('./data/day6-test.txt')}")
print(f"Part 2 - Real Data: {solve_part2('./data/day6-data.txt')}")
```
