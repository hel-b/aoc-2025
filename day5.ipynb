```python
# Part 1:
# data includes fresh ranges, a line separation, then, ingredient ids
# fresh range values are inclusive
# return number of fresh ingredients


def solve(fileName):
    # Read file
    fresh_ranges = []
    ingredients = []
    with open(fileName, "r") as f:
        r_tmp, i_tmp = f.read().split("\n\n")
        for r in r_tmp.split("\n"):
            if r:
                parts = r.split("-")
                fresh_ranges.append((int(parts[0]), int(parts[1])))
        ingredients = [int(i) for i in i_tmp.split("\n") if i]

    # Function to check if ingredient is fresh
    def is_fresh(ingredient):
        return any(low <= ingredient <= high for low, high in fresh_ranges)

    # Count fresh ingredients
    count = 0
    for i in ingredients:
        if is_fresh(i):
            count += 1
    return count


print(f"Part 1 - Test Data: {solve('./data/day5-test.txt')}")
print(f"Part 1 - Real Data: {solve('./data/day5-data.txt')}")
```

```python
# Part 2:
# Now they want to know number of valid fresh ingredient ids
def solve_part2(fileName):
    # Read file
    fresh_ranges = []
    with open(fileName, "r") as f:
        r_tmp, i_tmp = f.read().split("\n\n")
        for r in r_tmp.split("\n"):
            if r:
                parts = r.split("-")
                fresh_ranges.append((int(parts[0]), int(parts[1])))
    # Merge overlapping fresh ranges - first sort by start value
    fresh_ranges.sort(key=lambda v: v[0])
    # Add merged ranges to new list
    merged_ranges = [fresh_ranges[0]]
    for cur in fresh_ranges[1:]:
        prev = merged_ranges[-1]
        if cur[0] <= prev[1] and cur[1] > prev[1]:
            merged_ranges[-1] = (prev[0], cur[1])
        elif cur[0] > prev[1]:
            merged_ranges.append(cur)

    # Count valid fresh ingredients
    count = 0
    for i, f in merged_ranges:
        count += f - i + 1
    return count


print(f"Part 2 - Test Data: {solve_part2('./data/day5-test.txt')}")
print(f"Part 2 - Real Data: {solve_part2('./data/day5-data.txt')}")
#  449745556089941 is too high,
```
