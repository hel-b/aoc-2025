```python
# Part 1: given diagram of tachyon manifold, determine number of beam splitting operations
# Beam enter at "S" and travels downwards
# it splits at each "^" into two beams, left and right, until it exits


def solve_part1(file_path: str) -> int:
    rows = []
    with open(file_path, "r") as f:
        rows = f.read().strip().split("\n")
    beams = set()
    beams.add(rows[0].index("S"))
    splits = 0
    for row in rows[1:]:
        tmp_beams = set()
        for i in beams:
            if row[i] == "^":
                splits += 1
                tmp_beams.add(i - 1)
                tmp_beams.add(i + 1)
            else:
                tmp_beams.add(i)
        beams = tmp_beams
    return splits


print(f"Part 1 - Test Data: {solve_part1('./data/day7-test.txt')}")
print(f"Part 1 - Real Data: {solve_part1('./data/day7-data.txt')}")
```

```python
# Part 2:
# Now, the beam can only split in one direction, either left or right
# We need to determine the total possible number split sequences


def solve_part2(file_path: str) -> int:
    rows = []
    with open(file_path, "r") as f:
        rows = f.read().strip().split("\n")
    beams = [0] * len(rows[0])
    beams[rows[0].index("S")] = 1
    for row in rows[1:]:
        tmp_beams = [0] * len(rows[0])
        for i in range(len(beams)):
            if row[i] == "^":
                tmp_beams[i - 1] += beams[i]
                tmp_beams[i + 1] += beams[i]
            else:
                tmp_beams[i] += beams[i]
        beams = tmp_beams
    return sum(beams)


print(f"Part 2 - Test Data: {solve_part2('./data/day7-test.txt')}")
print(f"Part 2 - Real Data: {solve_part2('./data/day7-data.txt')}")
```
