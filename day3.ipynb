```python
# Part 1:
# Given list of battery ratings (one per line) in a file
# Objective: generate largest digit possible with two values
# from each rating ("joltage")
# VALUES MUST REMAIN IN THE SAME ORDER AS INPUT
# Example: 811111111111119 -> 89
# Return the sum of all largest digits from each rating


def solve(filename):
    ratings = []
    # Read in ratings from file
    with open(filename, "r") as f:
        ratings = f.read().strip().split("\n")
    joltages = []
    for rating in ratings:
        rating_values = [int(d) for d in rating]
        # Obtain the two largest digits while maintaining order
        digits = []
        digits.append(sorted(rating_values[:-1], reverse=True)[0])
        i = rating_values.index(digits[0]) + 1
        digits.append(sorted(rating_values[i:], reverse=True)[0])
        joltage = int("".join(map(str, digits)))
        joltages.append(joltage)
    return sum(joltages)


print(f"part1 - test: {solve('./data/day3-test.txt')}")
print(f"part1 - input: {solve('./data/day3-data.txt')}")
```

```python
# Part 2:
# Do the same thing but using 12 values from each joltage rating


def solve2(filename):
    ratings = []
    # Read in ratings from file
    with open(filename, "r") as f:
        ratings = f.read().strip().split("\n")
    joltages = []
    for rating in ratings:
        valArr = [int(d) for d in rating]
        # Obtain the first 12 largest digits while maintaining order
        # Using a variably sliding window depending on the number of digits left to select
        digits = []
        length = len(valArr)
        digits
        beg, end = 0, (length - 12) + 1
        while len(digits) < 12:
            digits.append(sorted(valArr[beg:end], reverse=True)[0])
            beg += valArr[beg:end].index(digits[-1]) + 1
            end = beg + (length - 12 + 1) - (beg - len(digits))
        joltage = int("".join(map(str, digits)))
        joltages.append(joltage)
    return sum(joltages)


print(f"part2 - test: {solve2('./data/day3-test.txt')}")
print(f"part2 - input: {solve2('./data/day3-data.txt')}")
```
