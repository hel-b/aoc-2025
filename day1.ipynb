```python
# Part 1:
# dial starts at 50
# dial runs from 0 to 99
# given a list of instructions
# R means turn right (increase number)
# L means turn left (decrease number)
# Count number of times the dial points to zero
def solve(filename):
    pos = 50
    count = 0
    # read lines from file
    lines = []
    with open(filename, "r") as f:
        lines = f.readlines()
    # adjust pos according to each line
    for line in lines:
        # strip return character
        line = line.strip()
        # extract first char and number
        direction = line[0]
        number = int(line[1:])
        # adjust pos
        pos += number if direction == "R" else -number
        # wrap around 0-99
        pos = (pos + 100) % 100
        # check if pos is zero
        if pos == 0:
            count += 1
    return count


print(f"part1 - test: {solve('test1.txt')}")
print(f"part1 - input: {solve('day1.txt')}")
```

```python
# Part 2:
# Now count any time the dial passes onto zero
# movements can be larger than 100 in either direction
def solve2(filename):
    pos = 50
    count = 0
    # read lines from file
    lines = []
    with open(filename, "r") as f:
        lines = f.readlines()
    # adjust pos according to each line
    for line in lines:
        # strip return character
        line = line.strip()
        # extract first char and number
        direction = line[0]
        number = int(line[1:])
        # store previous pos
        prev = pos
        # adjust pos without wrapping
        pos += number if direction == "R" else -number
        # count number of times pos passes onto zero (>=100 or <=0)
        if pos > 99:
            count += pos // 100
        elif pos <= 0:
            count += -pos // 100 + (1 if prev > 0 else 0)
        pos = ((pos % 100) + 100) % 100
    return count


print(f"part2 - test: {solve2('test1.txt')}")
print(f"part2 - input: {solve2('day1.txt')}")
```
